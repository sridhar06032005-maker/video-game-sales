```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv("d:/cleaned data/cleaned data vgsales.csv")
a.head()
```

```python
a.info()
```

```python
name = a["NA_Sales"].mean()
a["NA_Sales"]= a["NA_Sales"].apply(lambda x:name if x==0 else x)
print(a['NA_Sales'])
```

```python
name = a["EU_Sales"].mean()
a["EU_Sales"]= a["EU_Sales"].apply(lambda x:name if x==0 else x)
print(a['EU_Sales'])
```

```python
name = a["JP_Sales"].mean()
a["JP_Sales"]= a["JP_Sales"].apply(lambda x:name if x==0 else x)
print(a['JP_Sales'])
```

```python
name = a["Other_Sales"].mean()
a["Other_Sales"]= a["Other_Sales"].apply(lambda x:name if x==0 else x)
print(a['Other_Sales'])
```

```python
a.to_csv("d:/cleaned data vgsales.csv")
```

```python
a = pd.read_csv("d:/vgsales.csv")
a.head()
```

```python
x = a["Year"]
y = a["NA_Sales"]
plt.figure(figsize=(8, 6))
plt.title("vgsales")
plt.xlabel("NA_Sales")
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x = a["Year"]
y = a["EU_Sales"]
plt.figure(figsize=(8, 6))
plt.bar(x,y)
plt.show()
```

```python
x = a["Year"]
y = a["JP_Sales"]
plt.figure(figsize=(8, 6))
plt.hist(y)
plt.show()
```

```python
x = a["Year"]
y = a["Other_Sales"]
plt.figure(figsize=(8, 6))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Other_Sales',y='Other_Sales',data=a)
plt.title("Line Plot - Other Sales")
plt.xlabel("Index")
plt.ylabel("Other Sales")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="EU_Sales", y="EU_Sales", data=a)
plt.title("Bar Plot - EU_Sales")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```
